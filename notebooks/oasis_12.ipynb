```python
import pandas as pd
from sklearn.metrics import recall_score
from sklearn.metrics import precision_score
from sklearn.metrics import accuracy_score
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split

oasis1 = pd.read_csv("~/code/mkvph0ch/memobrain/raw_data/OASIS1/preprocessed_oasis1.csv")
oasis2 = pd.read_csv("~/code/mkvph0ch/memobrain/raw_data/OASIS2/preprocessed_oasis2.csv")
```

```python
X1=oasis1.drop(columns=['CDR'])
y1=oasis1['CDR']
X1
```

```python
X2=oasis2.drop(columns=['CDR'])
y2=oasis2['CDR']
X2
```

```python
frames=[oasis1, oasis2]
oasis12 = pd.concat(frames)
```

```python
oasis12.dtypes
```

```python
X12=oasis12.drop(columns=['CDR'])
y12=oasis12['CDR']
X12
```

Evaluating the merged dataset with the best estimator of the Oasis 2 analysis, training on the Oasis 2 dataset.

```python
best_model_oasis2 = RandomForestClassifier(n_estimators= 100, max_features= 'auto', max_depth= 8, criterion= 'gini')
best_model_oasis2.fit(X2, y2)
```

```python
y_pred_oasis2 = best_model_oasis2.predict(X12)
```

```python
accuracy_score(y12, y_pred_oasis2), recall_score(y12, y_pred_oasis2), precision_score(y12, y_pred_oasis2)
```

Evaluating the merged dataset with the best estimator of the Oasis 1 analysis, fitting in the Oasis 1 dataset.

```python
best_model_oasis1 = RandomForestClassifier(bootstrap=False, max_depth=20, max_features='sqrt', n_estimators=800)
best_model_oasis1.fit(X1, y1)
```

```python
y_pred_oasis1 = best_model_oasis1.predict(X12)
```

```python
accuracy_score(y12, y_pred_oasis1), recall_score(y12, y_pred_oasis1), precision_score(y12, y_pred_oasis1)
```

Evaluating the merged dataset with the best estimator of the Oasis 1 analysis, fitting in the merged dataset.

```python
X_train12, X_test12, y_train12, y_test12 = train_test_split(X12, y12, test_size=0.3, random_state=42)
```

```python
best_model_oasis1.fit(X_train12, y_train12)
y_pred12_1 = best_model_oasis1.predict(X_test12)
accuracy_score(y_test12, y_pred12_1), recall_score(y_test12, y_pred12_1), precision_score(y_test12, y_pred12_1)
```

Evaluating the merged dataset with the best estimator of the Oasis 2 analysis, fitting in the merged dataset.

```python
best_model_oasis2.fit(X_train12, y_train12)
y_pred12_2 = best_model_oasis2.predict(X_test12)
accuracy_score(y_test12, y_pred12_2), recall_score(y_test12, y_pred12_2), precision_score(y_test12, y_pred12_2)
```
