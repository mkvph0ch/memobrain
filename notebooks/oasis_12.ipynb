```python
import pandas as pd

oasis1 = pd.read_csv("~/code/mkvph0ch/memobrain/raw_data/OASIS1/preprocessed_oasis1.csv")
oasis2 = pd.read_csv("~/code/mkvph0ch/memobrain/raw_data/OASIS2/preprocessed_oasis2.csv")
```

```python
oasis1
```

```python
oasis2
```

```python
frames=[oasis1, oasis2]
oasis12 = pd.concat(frames)
```

```python
oasis12
```

```python
X=oasis12.drop(columns=['CDR'])
y=oasis12['CDR']
X
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
```

```python
from sklearn.svm import SVC
best_model = SVC(kernel="rbf", C= 5000, gamma=0.5)
best_model.fit(X_train, y_train)
```

```python
y_pred = best_model.predict(X_test)
```

```python
from sklearn.metrics import recall_score
recall_score(y_test, y_pred)
```
